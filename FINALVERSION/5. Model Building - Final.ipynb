# Formatting Data: Encoding Variables

Scikit's Random Forests only works on numerical values. Therefore, we will be converting categorical variables to binary dummy variables. For example, one column of our data might say "Is on Monday" and a 1 would be yes, while a 0 would be no. 

## Day of the Week

```python
from sklearn import preprocessing

train_days = traindf['DAY_OF_WEEK']
test_days = testdf['DAY_OF_WEEK']

# it is necessary to change the format of the days to [[1], [3], [7]...] rather than [1,3,7...] 
train_days = [[train_days[i]] for i in range(len(train_days))]
test_days = [[test_days[i]] for i in range(len(test_days))]

ohe = preprocessing.OneHotEncoder()

# Both the fitting and transform data must be in the form [[1], [3], [7]...]
ohe.fit([[i] for i in range(1,8)])

train_day_encoding = ohe.transform(train_days).toarray()
test_day_encoding = ohe.transform(test_days).toarray()

print train_day_encoding
print test_day_encoding
```

## Airline Carrier

```python
train_airline_id = traindf['AIRLINE_ID']
test_airline_id = testdf['AIRLINE_ID']

# LabelEncoder allows us to more easily turn categorical data into [1,2,3,4...] to then be used for OneHotEncoder
le = preprocessing.LabelEncoder()

# Extract all the unique items and fit to LabelEncoder 
airline_ids = list(set(train_airline_id))
le.fit(airline_ids)

# Converts airline id's : [13842, 34984, ...] to labels : [3, 5, 1 ...]
train_carrier_labels = le.transform(train_airline_id)
test_carrier_labels = le.transform(test_airline_id)

# Extract all the unique labels and change to the [[1], [2], [3]...] format for feeding into OneHotEncoder
carrier_list = list(set(train_carrier_labels))
carrier_list = [[carrier_list[i]] for i in range(len(carrier_list))]

# Using OneHotEncoder to fit the modified unique carrier_list
ohe2 = preprocessing.OneHotEncoder()
ohe2.fit(carrier_list)

# Convert format 
train_carrier_labels = [[train_carrier_labels[i]] for i in range(len(train_carrier_labels))]
test_carrier_labels = [[test_carrier_labels[i]] for i in range(len(test_carrier_labels))]

train_airline_encoding = ohe2.transform(train_carrier_labels).toarray()
test_airline_encoding = ohe2.transform(test_carrier_labels).toarray()

print train_airline_encoding
print test_airline_encoding
```

## Time of Day

Note: Random Forests can only be performed on at most 32 columns at once. We have 7 from the days of the week, and 17 from the carrier. Therefore, for now, we will treat time of day as a continuous variable. This okay because there are very few flights to/from BOS between 2 and 5 am. Therefore, I will translate the times to number of hours after 5 am. 

```python
test_time = testdf['DEP_TIME']
train_time = traindf['DEP_TIME']

def convertTime(time):
    for i in range(len(time)):
        old_time = time[i] 

        # If the time is before midnight, then simply subtract 500 from it 
        if old_time <= 2359 and old_time >= 500:
            new_time = old_time - 500
        # Otherwise, the time must be between 000 and 500, so simply add 2400 - 500 = 1900
        else: 
            new_time = old_time + 2400 - 500

        # Conversion from MMHH to "Number of Hours since 5 AM"
        time[i] = (new_time % 100)/ 60.0 + new_time / 100
    
    return time

test_time = convertTime(test_time)
train_time = convertTime(train_time)
```

```python
train_result = pd.DataFrame(train_time)

train_result = train_result.join(train_distance)
train_result = train_result.join(train_precipitation)
train_result = train_result.join(pd.DataFrame(train_day_encoding))

# Purpose: problems occur if you join data frames with the same column names (0,1,2,3...)
train_result.columns = ['DEP_TIME since 5am', 'DISTANCE', 'PRCP', 'MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT', 'SUN']
train_result = train_result.join(pd.DataFrame(train_airline_encoding))

train_result.head()
```

## Studying Number of Random ForestTrees

```python
X = train_result.as_matrix() 
Y = np.zeros(len(train_result))

# Temporarily using just 15 minutes to determine optimal number of trees
cutoff = 15 

# Converting arrival delay to binary
for i in range(len(train_result)):
    if traindf['ARR_DELAY'][i] >= cutoff: 
        Y[i] = 1
print Y

print "Proportion of 1's: %.3f" % (np.sum(Y) / len(Y))
```

```python
allscores = []

# Number of random forest decision trees
for k in [5,10,15,20]: 
    rfc = RandomForestClassifier(k)
    rfc = rfc.fit(X,Y)
    
    # Cross validation 5-fold
    score = cross_val_score(rfc, X, Y, cv=5)
    
    # append scores
    allscores.append(score)
```

```python
# Make boxplot for performance with different number of decision trees
# plt.figure(figsize=(20, 10))
sns.boxplot(allscores, names = ['5', '10', '15', '20'], color="pastel", widths=.5)
plt.title("Random Forest Performance with Cutoff=15 minutes")
plt.xlabel("Number of Decision Trees")
plt.ylabel("Classification Accuracy")
```

We will choose 10 trees as our optimal

## Studying the Delay Cutoff

We will study how our model compares to the trivial classifications (saying all flights are delayed or all flights are on-time). We want to pick the delay cutoff that gives the largest difference between our model's accuracy and that of the trivial classifiction. 

```python
# Number of Trees we found was optimal
numTrees = 10

# Storing all the Random Forest Classifiers
rfcs = []

# Range of values to be tested as the delay cutoff 
ranges = range(-15, 16, 5)

# Arrays to keep track of model and trivial scores
scores = [] 
trivial = []

X = train_result.as_matrix()

# Appens onto scores and trivial the appropriate cross_val score or the proportion of delays
def analyzeDelayCutoff(cutoff): 
    Y = np.zeros(len(train_result))

    # Creating a binary delay result
    for i in range(len(train_result)):
        if traindf['ARR_DELAY'][i] >= cutoff:
            Y[i] = 1.0

    
    rfc = RandomForestClassifier(numTrees)
    rfc = rfc.fit(X,Y)

    rfcs.append(rfc)
    
    # Trivial Accuracy 
    trivial.append(np.sum(Y)/len(Y))
    
    # Our Model's Accuracy 
    scores.append(np.mean(cross_val_score(rfc, X, Y, cv=5)))

for i in ranges:
    analyzeDelayCutoff(i)
```

```python
# Difference between trivial prediction 
differences = [scores[i] - max(trivial[i], 1 - trivial[i]) for i in range(len(ranges))]
plt.scatter(ranges, differences)

plt.title("Comparing the Predictions to that of the Trivial")
plt.xlabel("Delay Cutoff")
plt.ylabel("Difference in Trivial and Model Predictions")
```

It seems like the best cutoff is -5. And even in that case, we only have a ~5% gain on the trivial predictions

```python
plt.scatter(ranges, scores)
plt.xlabel("Cutoff")
plt.ylabel("Accuracy")
```

## Building a Random Forest Regression

We will also try using a Random Forest Regression, were instead of predicting a binary delay or no delay, we actually try to quantify how many minutes we think a flight will be delayed. 

```python
from sklearn.ensemble import RandomForestRegressor

Y = traindf['ARR_DELAY'].as_matrix()
X = train_result.as_matrix()

# We use 10 trees just like we did above 
regressor = RandomForestRegressor(10)
regressor = regressor.fit(X, Y)

predictions = regressor.predict(X)
```

```python
# Plotting the predicted delays vs. actual delays 
plt.scatter(predictions, Y, alpha = 0.01)
plt.xlabel("Predicted Delay")
plt.ylabel("Actual Delay")
plt.title("Comparing Predicted and Actual Delays")
plt.xlim(-200, 1000)
plt.ylim(-200, 1000)
```

```python
# Calculating the Mean Squared Error for the Training Set: 
mse = np.sqrt(np.mean([(Y[i] - predictions[i]) ** 2 for i in range(len(Y))]))

print 'On average, our prediction is off by %.1f minutes.' % mse
```
