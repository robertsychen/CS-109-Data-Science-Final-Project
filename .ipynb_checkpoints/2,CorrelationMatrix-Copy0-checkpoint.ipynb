# Initial Visualization: Correlation Matrix

Here is a preview of the relationships between the non-categorical variables of interest, visualized in a color correlation matrix. These features (which we'll investigate and visualize individually in the next notebook), are:

--Day of Week (as integer 1-7)

--Distance (from origin to destination) 

--Carrier

--Depature Time 

--Flight Duration 

--Arrival Delay 

--Day of Year

--and the response variable, delay time.

The other variables that we'll look at are the airport and the airline carrier, but we can't easily look at the correlations for those because they have many unrelated categories.



```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import StringIO

#EACH PERSON NEEDS THEIR OWN LINE HERE:
#Location = '/Users/robertchen/Documents/CS109AnnieEvanRobert-/ToFromBOSTrainingData.csv'

Location = 'C:\Users\Annie\Documents\GitHub\CS109AnnieEvanRobert-/ToFromBOSTrainingData.csv'
#Location = Evan's file path here

df = pd.read_csv(Location)
```

```python
print df.columns.tolist()
#Initial list of all variables
```

```python
df1=pd.DataFrame()
for reduced_variable in ['DAY_OF_WEEK','DISTANCE','CRS_DEP_TIME','ACTUAL_ELAPSED_TIME','ARR_DELAY']:
    df1[reduced_variable]=df[reduced_variable]
df1['CRS_DEP_TIME'] = (100 * (df1['CRS_DEP_TIME'] // 100)) + ((df1['CRS_DEP_TIME'] - (100* (df1['CRS_DEP_TIME'] // 100)))
                                                              * (100.0/60.0))
df1['DAY_OF_YEAR'] = 30*(df['MONTH'] - 1) + df['DAY_OF_MONTH']
print df1.columns.tolist()
```

```python
import seaborn as sns
```

```python
f, ax = plt.subplots(figsize=(18, 18))
sns.corrplot(df1, ax=ax)
```

The colors of bottom left represent correlations between each variable pair, and the top right represents the p values. 

The matrix highlights the strong correlation between flight duration (actual elapsed time) and distance. As we move forward with a model, it wouldn't make much sense to use both of them as predictive features. We also get a hint of a correlation between depature time and arrival delay, as well as departure time and arrival delay. 

We'll explore and visualize how all of these variables relate to arrival delay in the next notebook, but it's useful to see how they relate to each other first to be aware of intercorrelations.
