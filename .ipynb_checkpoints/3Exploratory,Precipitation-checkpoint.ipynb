More Exploratory Analysis

### How does amount of precipitation impact flight delays?

```python
precipitation = traindf['PRCP']
arr_delay = traindf['ARR_DELAY']

precipitation = [[precipitation[i]] for i in range(len(precipitation))]

from sklearn.linear_model import LinearRegression
lm = LinearRegression()

lm.fit(list(precipitation), list(arr_delay))
```

```python
plt.scatter(traindf['PRCP'], arr_delay, alpha = 0.005)
plt.xlabel('Amount of Precipitation (mm)')
plt.ylabel('ARR_DELAY')
plt.title("Precipitation vs. Arrival Delay")

plt.ylim(-30,30)
plt.xlim(0,700)

predictions = lm.predict(precipitation) 

plt.plot(traindf['PRCP'], predictions, color='red', linewidth=1)
plt.show()
```

```python
print lm.score(precipitation, arr_delay)
```

For a data set with 400,000 values and a lot of variance, such an $R^2$ score is not too bad!
