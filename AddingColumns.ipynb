```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
Location = '/Users/evanyao/CS109AnnieEvanRobert-/weather.csv'
file_location = '/Users/evanyao/CS109AnnieEvanRobert-/ToFromBOSTestData.csv'

df2 = pd.read_csv(Location)
df = pd.read_csv(file_location)

df.head()
```

```python
df2 = df2.drop_duplicates(subset = ['DATE'])
```

```python
print len(df2)
df2.head(10)
```

```python
for i in df2.index:
    dt = datetime.datetime.strptime(str(df2['DATE'][i]), '%Y%m%d')
    df2.loc[i, 'DATE'] = '{0}/{1}/{2:02}'.format(dt.month, dt.day, dt.year % 100)
```

```python
df2 = df2.set_index('DATE')
df2.head()
```

```python
d = df2.to_dict()
d = d['PRCP']
```

```python
prcp = [d[df['FL_DATE'][i]] for i in range(len(df))]
```

```python
df['PRCP'] = prcp
```
