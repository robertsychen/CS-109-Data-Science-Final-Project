# How does time of day for (scheduled) departure relate to arrival time delay?

```python
#7:15 AM is represented as 715, etc. We need to alter these times to be continuous. Example: change 715 to 725, to represent
#being one fourth of the way between 7:00 and 8:00. The formula: y = 100 * floor(x/100) + (x - 100*floor(x/100))*(100/60)
#We can do floor operation using integer division.

df11 = pd.DataFrame()
df11['CRS_DEP_TIME'] = (100 * (df['CRS_DEP_TIME'] // 100)) + ((df['CRS_DEP_TIME'] - (100* (df['CRS_DEP_TIME'] // 100)))
                                                              * (100.0/60.0))
df11['ARR_DELAY'] = df['ARR_DELAY']

plt.scatter(df11['CRS_DEP_TIME'], df11['ARR_DELAY'], alpha = 0.005)
plt.xlabel('CRS_DEP_TIME')
plt.ylabel('ARR_DELAY')

plt.xlim(0,2400)
plt.ylim(-100,200)

#In this case, it looks like flights with small magnitude of CRS_DEP_TIME are about the same in arrival delay. On the other
#hand, the large delay situations dramatically increase between time 1500 and 2000. This is likely to a snowball effect
#of flights being late from earlier flights in the day. It is worth studying both long delays and short delays here.

#Also, we won't even bother with the "all-data" regression, knowing it'll be pretty useless. We do a long delay regression and
#short delay regression.

#  Flight Time of Day Departing vs. Flight Delay
```

```python
plt.scatter(df11['CRS_DEP_TIME'], df11['ARR_DELAY'], alpha = 0.005)
plt.xlabel('CRS_DEP_TIME')
plt.ylabel('ARR_DELAY')

plt.xlim(500,2400)
plt.ylim(-30,30)

#Visualizing short delays. Doesn't appear to be much difference based on time of day. Notable: later times in the day
#are no more likely to be delayed a little. Earlier times in the day aren't any more likely to be early.

#  Flight Time of Day Departing vs. Flight Delay
```

```python
plt.scatter(df11['CRS_DEP_TIME'], df11['ARR_DELAY'], alpha = 0.05)
plt.xlabel('CRS_DEP_TIME')
plt.ylabel('ARR_DELAY')

plt.xlim(500,2400)
plt.ylim(30,210)

#Visualizing long delays. Between 1500 and 2000 time, peaking at 1800, it is most likely to have very long delays compared
#to other times. Also, it is most likely to have moderately long delays as well; the increase in density of flights in this
#time period is smooth/uniformly denser, both in the 30-90 minutes range and 90+ minute range.

#  Flight Time of Day Departing vs. Flight Delay
```

```python
#Regression on the shorter delays (-30 minutes to 30 minutes):

df12 = df11[df11['ARR_DELAY'] < 30]
df12 = df12[df12['ARR_DELAY'] > -30]

#run regression
df13 = np.asarray(df12)
data9 = df13[:,0]
data10 = df13[:,1]

#make data a 2-D numpy array
data9.resize(data9.shape[0],1)

lm5 = LinearRegression()
lm5.fit(data9, data10)
```

```python
print lm5.intercept_, lm5.coef_[0], lm5.score(data9, data10)

#Results of linear regression.
```

```python
plt.scatter(df11['CRS_DEP_TIME'], df11['ARR_DELAY'], alpha = 0.005)
plt.xlabel('CRS_DEP_TIME')
plt.ylabel('ARR_DELAY')

plt.xlim(500,2400)
plt.ylim(-30,30)

plt.plot(data9, lm5.predict(data9), color='red', linewidth=1)
plt.show()

#Graph of linear regression. Visually it's hard to tell initially, but there is a trend showing overestimation of how long
#it will take to arrive when flights are early in the day; this overly conservative estimate improves later in the day.

#  Flight Time of Day Departing vs. Flight Delay
```

```python
#Regression on the longer delays (30 minutes to 210 minutes):

df14 = df11[df11['ARR_DELAY'] > 30]
df14 = df14[df14['ARR_DELAY'] < 210]

#run regression
df15 = np.asarray(df14)
data11 = df15[:,0]
data12 = df15[:,1]

#make data a 2-D numpy array
data11.resize(data11.shape[0],1)

lm6 = LinearRegression()
lm6.fit(data11, data12)
```

```python
print lm6.intercept_, lm6.coef_[0], lm6.score(data11, data12)

#Results of linear regression. R=5.8% correlation.
```

```python
plt.scatter(df11['CRS_DEP_TIME'], df11['ARR_DELAY'], alpha = 0.05)
plt.xlabel('CRS_DEP_TIME')
plt.ylabel('ARR_DELAY')

plt.xlim(500,2400)
plt.ylim(30,210)

plt.plot(data11, lm6.predict(data11), color='red', linewidth=1)
plt.show()

#Graph of linear regression. This doesn't show much of the effect on long delays because there are so many more delays
#around 30 minutes that weigh down the model. Let's retry the model, working with only delays over 90 minutes. Also, the other
#problem is that the regression line can't take into account the high delay in the middle of the day but not either end,
#so adjust for that by removing time over 1900 for now.

#  Flight Time of Day Departing vs. Flight Delay
```

```python
#Regression on the longest delays (90 minutes to 210 minutes) and leaving before 1900 time:

df16 = df11[df11['ARR_DELAY'] > 90]
df16 = df16[df16['ARR_DELAY'] < 210]
df16 = df16[df16['CRS_DEP_TIME'] < 1900]

#run regression
df17 = np.asarray(df16)
data13 = df17[:,0]
data14 = df17[:,1]

#make data a 2-D numpy array
data13.resize(data13.shape[0],1)

lm7 = LinearRegression()
lm7.fit(data13, data14)
```
