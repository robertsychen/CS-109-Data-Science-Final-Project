# Introduction

In this project, we study what factors of a commercial airline flight influence how early or late it arrives, compared to its scheduled arrival time. Our first objective was to shed light on patterns within aspects of delayed flights. This could give officials within airline carriers and airport traffic/operation centers useful insight about how to time and schedule flights most efficiently. We also make predictions about how delayed specific flights will be, as well as more generally predict whether a single flight, given its particular characteristics, would tend to be more delayed or less delayed than other flights. Such models we developed would be interesting for prospective passengers who have numerous flight options and are hoping to avoid flight delays. The ideas for this project developed out of group members' overall interest in the transportation industry and infrastructure.

# Source of Data

Our primary data source was from one of the Bureau of Transportation Statistics' Transtats databases, which contains on-time arrival data for major carriers’ commercial domestic flights. Information in these data tables include departure and arrival delays, info about the flights such as flight number and carrier, as well as some data on causes for delays and cancelled/diverted flights. [http://www.transtats.bts.gov/DL_SelectFields.asp?Table_ID=236] Upon selecting a time window and choosing which of the 109 attributes one wishes to include in the table, one can download a .xls or .csv file with the information. 

We utilized data of domestic flights by major commercial airline carriers from October 2011 to September 2014, travelling between Boston (Logan Airport) and the 10 United States aiports with the most total traffic. The first two years of data constituted the training set (size 426112), while the third year of data constituted the test set (size 205961). In our table, 48 characteristics of each flight are given, including its date, duration, time of departure, arrival delay and airline carrier. An additional column (not from the BTS) called precipitation was added. That data was acquired through the National Climatic Data Center.

```python
#We used this code after we manually obtained data for each month, in order to merge the csv files together.

import pandas as pd
import glob

path =r'C:\Users\Annie\Documents\GitHub\CS109AnnieEvanRobert-\Data'
allfiles = glob.glob(path + "/*.csv")
frame=pd.DataFrame()
lst=[]
for files in allfiles:
    iter_csv=pd.read_csv(files,iterator=True,chunksize=1000)
    df = pd.concat([chunk[chunk['ORIGIN']=='BOS'] for chunk in iter_csv])
    lst.append(df)
    print len(lst)
    iter_csv_1=pd.read_csv(files,iterator=True,chunksize=1000)
    df1 = pd.concat([chunk[chunk['DEST']=='BOS'] for chunk in iter_csv_1])
    lst.append(df1)
    print len(lst)
    
frame = pd.concat(lst,ignore_index=True)

frame.to_csv('Flights To and From Boston past 3 years.csv')
```
